```python
import os  
import math
from PIL import Image
from fpdf import FPDF
pdf = FPDF()
import re

sdir = "Baskı/"                 ## sdir değişkeni: RESİMLERİN OLDUĞU KLASÖR YOLU, JUPYTER NOTEBOOKS klasörünün içine "Baskı"
                                 ## diye bir dosya oluşturup pdfe çevireceğiniz dosyaları oraya atın
  

from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(sdir) if isfile(join(sdir, f))]

pdf = FPDF()         #Resimler için oluşturduğum obje
pdf1 = FPDF()        #Kodlar için oluşturduğum obje



for i2 in range(math.ceil((len(onlyfiles)/66))):          #Her 66 png de bir yeni sayfaya geç
    print("Sayfa: " , i2+1)
    pdf.add_page()
    pdf1.set_font('Arial', 'B', 8)        #Font değeri değiştirilebilir
    pdf1.add_page()
    
    onlyfilesSayfa = onlyfiles[(i2)*66:(i2+1)*66]
    
    for i in range(1, 7):
        if(i>len(onlyfilesSayfa)) : break
        
        fname =sdir + onlyfilesSayfa[i-1]          
        
        if os.path.exists(fname):                   

            image = fname
            pdf.image(image,i*25,10,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])   #Direkt olarak kod halinde yazdırmak için dosya isimlerinin .png kısmını almıyorum 
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i)
    pdf1.ln(20)

    for i in range(7, 13):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1]  
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,35,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i) 
        
    pdf1.ln(20)

    for i in range(13, 19):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,60,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i)
        
    pdf1.ln(20)

    for i in range(19, 25):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,85,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i) 
        
    pdf1.ln(20)

    for i in range(25, 31):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,110,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i) 
        
    pdf1.ln(20)

    for i in range(31, 37):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,135,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i) 
        
    pdf1.ln(20)

    for i in range(37, 43):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,160,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i) 
        
    pdf1.ln(20)

    for i in range(43, 49):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,185,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i) 
        
    pdf1.ln(20)

    for i in range(49, 55):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,210,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i)
        
    pdf1.ln(20)

    for i in range(55, 61):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,235,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i) 
        
    pdf1.ln(20)

    for i in range(61, 67):
        if(i>len(onlyfilesSayfa)) : break
        fname = sdir + onlyfilesSayfa[i-1] 
        if os.path.exists(fname):      
            image = fname
            pdf.image(image,25+(i%6)*25,260,17,17)
            pdf1.cell(30, 40,onlyfilesSayfa[i-1][:-4])
        else:
            print("Dosya bulunamadı:", fname)
        print("eklendi %d" % i) 
        
    
    
    
   
pdf.output("BASKI.pdf", "F")
pdf1.output('KODLAR.pdf', 'F')
print("Bu işlem sonunda BASKI adında bir pdf dosyası ve aynı sırada kodları içeren KODLAR pdfi oluşturulacaktır")
print("işlem tamam")
```
